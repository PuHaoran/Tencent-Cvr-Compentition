```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
```

```python
def read_csv_file(f, logging=False):
    """ 读取csv文件 """
    data = pd.read_csv(f)
    if logging:
        print(data.head())
        print('包含以下列：')
        print(data.columns.values)
        print(data.describe())
        print(data.info())
    return data

# 根据已有数据构造更多特征
def process_app_categories_first_class(category):
    """ app_category第一类编码 """
    category = str(category)
    if len(category) == 1:
        if int(category) == 0:
            return 0
    else:
        return int(category[0])

def process_app_categories_second_class(category):
    """ app_category第二类编码 """
    category = str(category)
    if len(category) < 3:
        return 0
    else:
        return int(category[1:])
    
def process_age(age):
    """ 处理年龄 """
    age = int(age)
    if age == 0:
        return 0
    elif age < 15:
        return 1
    elif age < 25:
        return 2
    elif age < 40:
        return 3
    elif age < 60:
        return 4
    else:
        return 5

def process_province(hometown):
    """ 处理省份 """
    hometown = str(hometown)
    province = int(hometown[0:2])
    return province

def process_city(hometown):
    """ 处理城市 """
    hometown = str(hometown)
    if len(hometown) > 1:
        province = int(hometown[2:])
    else:
        province = 0
    return province

def process_get_time_day(t):
    """ 获取时间DD """
    t = str(t)
    t = int(t[0:2])
    return t

def process_get_time_hour(t):
    """ 获取时间HH """
    t = str(t)
    t = int(t[2:4])
    if t<6:
        return 0
    elif t<12:
        return 1
    elif t<18:
        return 2
    else:
        return 3
```

### 读取数据

```python
train = read_csv_file('../data/pre/train.csv', logging=True)
test = read_csv_file('../data/pre/test.csv', logging=True)
ad = read_csv_file('../data/pre/ad.csv', logging=True)
```

### 特征工程

```python
# appID  appCategory APP分类
app_categories = read_csv_file('../data/pre/app_categories.csv', logging=True)
app_categories['app_categories_first_class'] = app_categories['appCategory'].apply(process_app_categories_first_class)
app_categories['app_categories_second_class'] = app_categories['appCategory'].apply(process_app_categories_second_class)
```

```python
user = read_csv_file('../data/pre/user.csv', logging=False)
user.columns
```

简单统计下user表信息，比如根据user.age的分布获取age的扩展特征。

```python
user[user.age!=0].describe()
```

```python
user.age.value_counts().head(20)
```

获取user的扩展特征
    - hometown 用户出生地，千位百位表示省份，十位个位表示省内城市
    - residence 常住地，编码方式与hometown相同

```python
user['process_age'] = user['age'].apply(process_age)
user['process_hometown_province'] = user['hometown'].apply(process_province)
user['process_hometown_city'] = user['hometown'].apply(process_city)
user['process_residence_province'] = user['residence'].apply(process_province)
user['process_residence_city'] = user['residence'].apply(process_city)
```

clickTime格式DDHHMM，这里通过切分时间获取天、时的扩展特征。

```python
train['process_clickTime_day'] = train['clickTime'].apply(process_get_time_day)
train['process_clickTime_hour'] = train['clickTime'].apply(process_get_time_hour)
test['process_clickTime_day'] = test['clickTime'].apply(process_get_time_day)
test['process_clickTime_hour'] = test['clickTime'].apply(process_get_time_hour)
```

```python
# 合并数据集
train_user = pd.merge(train, user, on='userID')
train_user_ad = pd.merge(train_user, ad, on='creativeID')
train_user_ad_app = pd.merge(train_user_ad, app_categories, on='appID')
```

取出数据和label，为下一步放入模型进行训练做准备。

```python
features_labels = ['creativeID','userID','positionID',
 'connectionType','telecomsOperator','process_clickTime_day','process_clickTime_hour','age', 'gender' ,'education',
 'marriageStatus' ,'haveBaby' , 'residence' ,'process_age',
 'process_hometown_province', 'process_hometown_city','process_residence_province', 'process_residence_city',
 'adID', 'camgaignID', 'advertiserID', 'appID' ,'appPlatform' ,
 'app_categories_first_class' ,'app_categories_second_class']
```

```python
x_user_ad_app = train_user_ad_app.loc[:, features_labels][60000:75000]
x_user_ad_app = x_user_ad_app.values
x_user_ad_app = np.array(x_user_ad_app,dtype='int32')
y_user_ad_app = train_user_ad_app.loc[:,['label']].values[60000:75000]
```

### 随机森林&&特征重要度排序

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score, train_test_split
```

```python
random_forest = RandomForestClassifier(n_estimators=50,
                                random_state=0,
                                n_jobs=-1)
random_forest.fit(x_user_ad_app, y_user_ad_app.reshape(y_user_ad_app.shape[0],))
```

```python
importances = random_forest.feature_importances_
indices = np.argsort(importances)[::-1]
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

for f in range(x_user_ad_app.shape[1]):
    print("%2d) %-*s %f" % (f + 1, 30, 
                            np.array(features_labels[indices[f]]), 
                            importances[indices[f]]))
```

```python
plt.title('Feature Importances')
plt.bar(range(x_user_ad_app.shape[1]), 
        importances[indices],
        color='lightblue', 
        align='center')

plt.xticks(range(x_user_ad_app.shape[1]), 
           np.array(features_labels)[indices], rotation=90)
plt.xlim([-1, x_user_ad_app.shape[1]])
plt.tight_layout()
#plt.savefig('./random_forest.png', dpi=300)
plt.show()
```

### 随机森林调参

如果我們使用 sklearn.metrics.log_loss 獨立，換句話說，log_loss(y_true,y_pred), 生成一個正數--越小，性能就越好。

但是，如果我們使用 'neg_log_loss' 作為得分方案，則得分為負，得分越大，性能就越好。

```python
sum(y_user_ad_app.reshape(y_user_ad_app.shape[0],))
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier
param_grid = {
    'n_estimators': [10, 50, 100, 200],
    'max_features': [0.6, 0.7, 0.8]
}

rf = RandomForestClassifier()
rfc = GridSearchCV(rf, param_grid, scoring='neg_log_loss', cv=3, n_jobs=2)
rfc.fit(x_user_ad_app, y_user_ad_app.reshape(y_user_ad_app.shape[0],))
print(rfc.best_score_)
print(rfc.best_params_)
```

```python
# 查看正负样本比
len(train_user_ad_app[train_user_ad_app['label'] == 1]) / len(train_user_ad_app[train_user_ad_app['label'] == 0])
```

可以看出正负样本数量相差很大，数据严重不均衡。这里我们采用bagging进行修正。

```python
from blagging import BlaggingClassifier
```

```python
# 处理unbalanced数据的classifier
classifier = BlaggingClassifier(n_jobs=-1)
classifier.fit(x_user_ad_app, y_user_ad_app)
```
